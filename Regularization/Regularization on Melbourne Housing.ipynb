```python
import pandas as pd
df = pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df['Price'] = df.Price.fillna(df['Price'].mean())
```

```python
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[cols_to_use]
```

```python
df.head()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
```

```python
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.head()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first = True)
df.head()
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

### Using L1 (Lasso) Regularization model

```python
from sklearn.linear_model import Lasso
Lasso_reg = Lasso(alpha=50, max_iter=100, tol=0.1).fit(X_train, y_train)
```

```python
Lasso_reg.score(X_test, y_test)
```

### Using L2 (Ridge) Regularization model

```python
from sklearn.linear_model import Ridge
Ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1).fit(X_train, y_train)
```

```python
Ridge_reg.score(X_test, y_test)
```

##### LOL: In this example L1 and L2 doesn't make any huge difference but it's very helpful to reduce underfitting or overfitting
